```python
#สร้างข้อมูลบน python

data = {
    "location":['CityA','CityB'],
    "Temperature":['Predict','Actual'],
    "Jan-2010":[30,32],
    "Feb-2010":[45,43],
    "Mar-2010":[24,22]
}

data
```

```python
df = pd.DataFrame(data, columns=['location','Temperature','Jan-2010','Feb-2010','Mar-2010'])
df
```

```python
df2 = pd.melt(df)
df2
```

```python
df2_melt = pd.melt(df, id_vars =['location','Temperature'], var_name='Month_year',value_name ='Sum')
df2_melt
```

```python
df2 =df2_melt[['Month_year','Temperature','location','Sum']]
df2
```

```python
#เรียง sum จากน้อยไปมาก
df2.sort_values('Sum')
```

```python
#เรียง sum จากมากไปน้อย
df2.sort_values('Sum',ascending=False)
```

```python
df2 = df2.rename(columns = {'Month_year':'Month_Y'})
df2
```

```python
#เลือกผู้ที่มี Orderquantity > 28 ขึ้นไป
sale[sale.OrderQuantity > 28][['OrderID','OrderQuantity']]
```

```python
#เลือกผู้ที่มี Orderquantity > 28 ขึ้นไป
# Sales >2000
# Discount < 0.09

sale[(sale.OrderQuantity > 28)&(sale.Sales > 2000)&(sale.Discount < 0.09)][['OrderID','OrderQuantity','Sales','Discount']]
```
